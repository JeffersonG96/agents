### Document Loaders
Permite cargar documentos que incluye tanto el contenido del documento como su metadata

Tambien permite carga informacion online

Existe mas de 1000 tipos de document loaders (diferentes formatos)

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("D:\\Cursos\\Certificados\\CV2_Completo_Jefferson_Guaman.pdf")

texto = loader.load()
```

```python
for i, page in enumerate(texto):
    print(f"contenido: {page.metadata}")
    print(f"contenido: {page.page_content}")
```

```python
from langchain_community.document_loaders import WebBaseLoader
loader_web = WebBaseLoader("https://www.nvidia.com/en-us/glossary/generative-ai/")

docs = loader_web.load()

# for i, page in enumerate(docs):
#     print(f"contenido: {page.page_content}")
```

#### Dividir el texto en chunks mas pequeños 
Text Splitter permite configurar el tamaño del chunk y los caracteres q se interponen con el anterir chunk (división intelligente)

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
tex_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 10000,
    chunk_overlap = 200
)

#carga la división en chunks
chunks = tex_splitter.split_documents(docs)
len(chunks)
```

#### Embeddings
Representaciones vectoriales numéricas de datos (texto, imágenes, audio) en un espacio multidimensional, donde la distancia entre vectores refleja similitud semántica.

```python
from  langchain_openai import OpenAIEmbeddings
import numpy as np

embeddings = OpenAIEmbeddings(model="text-embedding-3-large")

texto1 = "La capital de Ecuador es Quito"
texto2 = "Quito es la capital de Ecuador"

vec1 = embeddings.embed_query(texto1)
vec2 = embeddings.embed_query(texto2)

print(f"Dimensión de los vectores: {len(vec1)}")

cos_sim = np.dot(vec1,vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2) ) 
#1 = muy similares, 0 = no relacionados, -1 = opuestos.
print(f"similitud coseno entre vec1 y vec2: {cos_sim: .3f}")
```
