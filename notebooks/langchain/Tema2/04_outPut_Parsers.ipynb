## OutPut Parsers

Convertir la salida “libre” del modelo en datos estructurados, validados y seguros, listos para usarse en código real.

```python
#Introduccion pydantic
from pydantic import BaseModel, Field

class Usuario(BaseModel):
    id: int
    nombre: str
    activo: bool = True

data = {"id": "123","nombre":"Jefferson"}

usuario = Usuario(**data)
print(usuario.model_dump_json)
```

```python
from langchain_openai import ChatOpenAI

class AnalisisTexto(BaseModel):
    resumen: str = Field(descripcion="Resumen breve del texto.")
    sentimiento: str = Field(descripcion="sentimiento del texto (positico, neutroto o negativa)")

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.5)

structured_llm = llm.with_structured_output(AnalisisTexto)

texto_prueba = "Me encanto la nueva película de avata, tiene mucho efectos especiales"

resultado = structured_llm.invoke(f"Analiza el siguiente texto: {texto_prueba}")

resultado
```

```python
resultado.model_dump_json()
```
